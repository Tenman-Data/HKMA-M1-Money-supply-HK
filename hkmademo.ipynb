```python

import urllib.request

url = 'https://api.hkma.gov.hk/public/market-data-and-statistics/monthly-statistical-bulletin/money/supply-components-all?pagesize=100000'

with urllib.request.urlopen (url) as req:

    x=req.read()
```

```python
import json

y = json.loads(x)

print(type(y["result"]))
print(y["result"])
```

```python
result = y["result"]
records = result["records"]
records
# type(records)
```

```python

records[0]
```

```python
import pandas as pd 
df = pd.DataFrame.from_dict(records)
```

```python
df.head(5)
```

```python
type(df["end_of_month"][0])
```

```python
import matplotlib 
import matplotlib.pyplot as plt

df.plot(kind='line', x='end_of_month', y='m1_supply' , color = 'red')
plt.show()
```

```python

ax=plt.gca()
df.plot(kind='line', x='end_of_month', y='m1_supply' , color = 'red' ,ax=ax)
df.plot(kind='line', x='end_of_month', y='m2_supply' , color = 'green', ax= ax)
plt.show()
```

```python

df2 = df.sort_values(by=['end_of_month'])
df2 = df2.reset_index(drop=True)
print(df2)
```

```python

ax=plt.gca()
df2.plot(figsize=(20,10), kind='line', x='end_of_month', y='m1_supply' , color = 'red' ,ax=ax)
df2.plot(kind='line', x='end_of_month', y='m2_supply' , color = 'green', ax= ax)
plt.title("M1 M2")
plt.xlabel("Year Month")
plt.ylabel("Mil HKD")
plt.show()
```

```python
import datetime as dt

pd.to_datetime(df2['end_of_month'], format="%Y-%m-")
```

```python
df3 = df2.copy()

df3['end_of_month']=pd.to_datetime(df3['end_of_month'], format="%Y-%m-")

df3['end_of_month'][0]
```

```python

ax=plt.gca()
df3.plot(figsize=(20,10), kind='line', x='end_of_month', y='m1_supply' , color = 'red' ,ax=ax)
df3.plot(kind='line', x='end_of_month', y='m2_supply' , color = 'green', ax= ax)
plt.title("M1 M2")
plt.xlabel("Year Month")
plt.ylabel("Mil HKD")
plt.show()
```

```python

firstdate  = df3["end_of_month"][0]
firstdate 

dftest = df3["end_of_month"] - firstdate
type(dftest)


df4 = df3.copy()
df4.info()
daydiff = df4["end_of_month"] - firstdate
type(daydiff[0])
df4["daydiff"]=daydiff.dt.days
```

```python

df4.head(5)
df4.info()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = df4['daydiff'].values.reshape(-1,1)
Y = df4['m1_supply'].values.reshape(-1,1)


```

```python

linear_regressor = LinearRegression()
linear_regressor.fit(X,Y)
```

```python
Y_pred = linear_regressor.predict(X)
Y_pred
```

```python

plt.figure(figsize=(20,10))
ax=plt.gca()
plt.plot(df4['end_of_month'],df4['m1_supply'], color = 'red')
plt.plot(df4['end_of_month'],Y_pred, color = 'grey')
plt.show()
```

```python
roll_mean=df4['m1_supply'].rolling(window=20).mean()
roll_std=df4['m1_supply'].rolling(window=20).std()
```

```python


plt.figure(figsize=(20,10))
ax=plt.gca()
plt.plot(df4['end_of_month'],df4['m1_supply'], color = 'red')
plt.plot(df4['end_of_month'],Y_pred, color = 'grey')
plt.plot(df4['end_of_month'],roll_mean, color = 'blue')

plt.show()


```

```python
plt.figure(figsize=(20,10))
plt.plot(df4['end_of_month'],roll_std, color = 'blue')

plt.show()
```

```python


from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.rcParams.update({'figure.figsize':(9,7) , 'figure.dpi':120})

fig, axes = plt.subplots(3,2, sharex=True)
axes[0,0].plot(df4['m1_supply']);axes[0,0].set_title('orgi')
plot_acf(df4['m1_supply'], ax=axes[0,1])
                                                     
axes[1, 0].plot(df4["m1_supply"].diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(df4["m1_supply"].diff().dropna(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(df4["m1_supply"].diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(df4["m1_supply"].diff().diff().dropna(), ax=axes[2, 1])

plt.show()                                               
```
